# SELECTED Variables - by Correlation

```python
# Select only the independent variables for correlation matrix
independent_variables = merged_df[[column for column in merged_df.columns if column not in ['YEAR', 'TCI_standardized']]]

# Calculate the correlation matrix
correlation_matrix = independent_variables.corr()

correlation_matrix
```

```python
# Use seaborn to create a heatmap
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Variable Correlation')
plt.show()
```

```python
# Initialize a dictionary to count low correlations for each variable
low_correlation_counts = {col: 0 for col in correlation_matrix.columns}

# Define the correlation threshold
threshold = 0.7

# Priority variables to keep based on subject importance (CPI is a must-keep if possible)
priority_keep = ['CPI_standardized']

# List of GDP variables among which we need to keep at least one
gdp_variables = ['GDP_standardized', 'GDP_Santiago_standardized']

# Select which GDP variable to keep (preferring the one with the lowest max correlation with the priority keep variable)
min_corr_gdp = min(gdp_variables, key=lambda x: max(abs(correlation_matrix.loc[x, priority_keep])))

# Initialize list to keep variables, starting with the selected GDP variable
variables_to_keep = [min_corr_gdp] + priority_keep

# Check other variables and add them if they are not highly correlated with the already selected ones
for var in correlation_matrix.columns:
    if var not in variables_to_keep:
        if all(abs(correlation_matrix.loc[var, kept]) < threshold for kept in variables_to_keep):
            variables_to_keep.append(var)

# Final check to remove any internally highly correlated variables (if any slipped through)
final_variables_to_keep = []
for var in variables_to_keep:
    if all(abs(correlation_matrix.loc[var, other]) < threshold for other in variables_to_keep if other != var):
        final_variables_to_keep.append(var)

print("Variables to potentially keep after thorough checks:")
print(final_variables_to_keep)
```

```python
# Set Default% as your Y variable
y_variable = 'Default%'

# Extract the correlations of the Y variable with other variables
y_correlations = correlation_matrix.loc[y_variable]

# Sort the correlations in descending order to prioritize variables highly correlated with Y
sorted_y_correlations = y_correlations.abs().sort_values(ascending=False)

# Initialize an empty list to store selected variables
selected_variables = []

for variable in sorted_y_correlations.index:
    if variable == y_variable:
        continue  # Skip the Y variable to avoid selecting it as a predictor
    # Check if the variable is highly correlated with any already selected variables
    if not any(abs(correlation_matrix.loc[selected_var, variable]) > 0.5 for selected_var in selected_variables):
        # Add the variable to the selected list
        selected_variables.append(variable)

# Print the selected variables
print("Selected Variables:")
print(selected_variables)
```
